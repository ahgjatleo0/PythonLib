### 데이터를 시각화하는 라이브러리

#### 시각화 라이브러리(모듈) 불러오기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### line plot : 선 그래프


```python
# 선 그래프 함수 : plot()
y = [2, 4, 6, 8]
plt.plot(y)
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
plt.plot(x,y)
plt.show()
```

#### 그래프 스타일 지정


```python
# 선 스타일
y = [2, 4, 6, 8]
plt.plot(y ,ls = ':') # ls : line style
plt.show()
```

#### 그래프 값 마킹하기


```python
# 마커포인트 : marker
y = [2, 4, 6, 8]
plt.plot(y ,marker = '.') # ls : line style
plt.show()
```

```python
# 선 두께 : line width ==> lw
y = [2, 4, 6, 8]
plt.plot(y ,lw = 3, marker = 'o') # ls : line style
plt.show()
```

```python
# 선 색상 : line color ==> c
y = [2, 4, 6, 8]
plt.plot(y ,lw = 3, marker = 'o', c = 'lightgray' ) # ls : line style
plt.show()
```

```python
# marker 색상 지정하기
y = [2, 4, 6, 8]
plt.plot(y ,lw = 3, marker = 'o', c = 'lightgray', mfc = 'gray' ) 
# mfc : marker 색상
plt.show()
```

#### 실습


```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]
plt.plot(x,y ,lw = 3, ls = ':', marker = 'o', c = 'lightgray', mfc = 'black' ) 
plt.show()
```

```python
# 그래프 사이즈 지정
plt.figure(figsize = (5,3))
plt.plot(x,y ,lw = 3, ls = ':', marker = 'o', c = 'lightgray', mfc = 'black' ) 
plt.show()
```

#### 그래프 옵션 설정

```python
# 그림 범위 지정
plt.plot(x,y ,lw = 3, ls = ':', marker = 'o', c = 'lightgray', mfc = 'black' ) 
plt.xlim(-2, 10)  # x축의 범위 제한
plt.ylim(2, 8)  # y축의 범위 제한
plt.show()

# 만약 2 ~ 8 사이의 값을 보고 싶을 때 (zoomin) 
```

```python
# ticks : x,y 축 표시값 설정
plt.plot(x,y ,lw = 3, ls = ':', marker = 'o', c = 'lightgray', mfc = 'black' ) 
plt.xticks([0, 3, 6])
plt.yticks([1, 5, 9])
plt.show()
```

```python
# 눈금선 표시 : gird()
plt.plot(x,y ,lw = 3, ls = ':', marker = 'o', c = 'lightgray', mfc = 'black' ) 
plt.grid() # grid 함수 호출
plt.show()
```

```python
# 여러개의 그래프 출력하기
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
z = [3, 6, 9, 12]
plt.plot(x, label = 'x')
plt.plot(y, label = 'y')
plt.plot(z, label = 'z')
plt.legend(loc = 'best') # 범례표시 기본값 : best
plt.xlabel('x') # x축의 이름
plt.ylabel('y') # y축의 이름
plt.title('my graph') # 그래프 제목
plt.show() # x, y, z 모두 출력
```

#### 요일별 교통사고 시각화 실습

##### 교통사고 사망자 수 count

```python
data.head(0)
```

```python
# value_counts() 특정 컬럼의 값들의 개수를 각각 출력
# 하나의 컬럼안에(시리즈에서) 있는 유니크 값(중복없는)들과 그 값들의 개수를 각각 출력
# 시리즈에서만 사용 가능하다!!
data['사망자수'].value_counts()
```

- 위 결과로 알 수 있는 점 : 우리의 데이터는 사망자가 1인인 사고가 교통사고의 대부분을 차지하고 있다

#### 요일별 사고 건수 count
- 요일별 사고건수 확인
- 요일 순서대로(월~일) 출력

```python
# 1. 요일 컬럼 접근
data['요일']
```

```python
# 2. 중복없이 값 확인(월 ~ 일)
# unique() : 중복제거
data['요일'].unique()
```

```python
# 3. 요일별 사고 건수 확인
y = data['요일'].value_counts()
y
```

```python
# 4. 요일 순서대로 변경
# 행과 열로 이루어진 2차원 형태로 출력하기 위해 대괄호 2개를 사용하여 차원을 2차원으로 변경
y = y[['월', '화', '수', '목', '금', '토', '일']]
y
```

```python
# 5. bar 차트 그리기
x = y.index # x가 없으므로 요일 정보를 출력하기 위해 인덱스명을 추출
plt.bar(x,y)
plt.ylim(500, 620) # y축에 범위 저장

plt.xlabel('요일')  # x축에 이름 저장
plt.ylabel('사고 건수') # y축에 이름 저장
plt.title('요일별 사망 교통사고 건수') # 전체 이름(title) 저장

# ticks() : 데이터가 어디서부터 시작되는지 정확한 지점을 볼 수 있다.
plt.yticks(y) # y의 정확한 값들을 표현
# xticks(x) ==> 자동적으로 부여가 된다.

plt.show() # show()는 항상 마지막에 써주자!
```
